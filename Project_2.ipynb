```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
import os
```

```python
os.listdir("Data/")
```

```python
us_only = pd.read_csv('Data/title-akas-us-only.csv', low_memory=False)
us_only.head()
```

```python
basics = pd.read_csv("Data/title.basics.tsv.gz", sep='\t', low_memory=False)
basics.info()
basics.head()
```

```python
basics.shape
```

```python
filter_us_titles = basics['tconst'].isin(us_only['titleId'])
basics = basics[filter_us_titles]
basics.head()
```

```python
basics.shape
```

```python
basics = basics.replace({'\\N':np.nan})
basics.isna().sum()
```

```python
basics.dropna(subset=['runtimeMinutes'], inplace=True)
basics.dropna(subset=['genres'], inplace=True)
```

```python
basics.isna().sum()
```

```python
filter_movie = basics['titleType'].str.contains('movie')
basics = basics.loc[filter_movie]
basics.head()
```

```python
basics['startYear'].dtypes
```

```python
basics['startYear'] = basics['startYear'].astype(float)
basics['startYear'].dtypes
```

```python
basics = basics.loc[(basics['startYear'] >= 2000)&(basics['startYear'] <= 2022)]
basics.head()
```

```python
filter_documentaries = basics['genres'].str.contains('Documentary')
basics = basics[~filter_documentaries]
basics.head()
```

```python
basics.info()
```

```python
basics_clean = 'Data/title.basics.csv'
basics.to_csv(basics_clean, index=True)
```

```python
ratings = pd.read_csv('Data/title.ratings.tsv.gz', sep='\t', low_memory=False)
ratings.info()
ratings.head()
```

```python
filter_basics = ratings['tconst'].isin(basics['tconst'])
ratings = ratings[filter_basics]
```

```python
ratings = ratings.replace({'\\N':np.nan})
```

```python
ratings.info()
ratings.head()
```

```python
ratings_clean = 'Data/title.ratings.csv'
ratings.to_csv(ratings_clean, index=True)
```
